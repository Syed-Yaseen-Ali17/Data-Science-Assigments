```python
# 1. Data Preprocessing:
import pandas as pd
df = pd.read_csv('anime.csv')
df  
```

```python
# Handle missing values
df.dropna(inplace=True) 
```

```python
# Impute null values with mode
for col in df.columns:
  df[col].fillna(df[col].mode()[0], inplace=True)
df
```

```python
# Explore the dataset
df.head()
```

```python
df.info() 
```

```python
df.describe() 
```

```python
# 2. Feature Extraction:
# Convert categorical features to numerical using Label Encoding
X_cate = df[['name','genre','type','episodes']]

from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()

X_cate['name']=LE.fit_transform(X_cate['name'])
X_cate['genre']=LE.fit_transform(X_cate['genre'])
X_cate['type']=LE.fit_transform(X_cate['type'])
X_cate['episodes']=LE.fit_transform(X_cate['episodes'])

X_cate.head()
```

```python
#Normalization 
X_cont = df[['anime_id','members','rating']]

from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()
X_cont_MM = MM.fit_transform(X_cont)
X_cont_MM = pd.DataFrame(X_cont_MM)
X_cont_MM.columns = list(X_cont)
X_cont_MM.head() 
```

```python
# 3. Recommendation System:
# Design a function to recommend anime based on cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(anime_name, df, top_n=10, similarity_threshold=0.5):
  # Prepare data for cosine similarity
  X_cos = X_cate.values  # Use the encoded categorical features

  # Calculate cosine similarity matrix
  cosine_sim = cosine_similarity(X_cos)

  # Get index of the anime
  anime_index = df[df['name'] == anime_name].index[0]

  # Get similarity scores for the anime
  similarity_scores = list(enumerate(cosine_sim[anime_index]))

  # Sort anime by similarity score
  sorted_anime = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

  # Get recommendations above the similarity threshold
  recommended_anime = []
  for i in range(1, len(sorted_anime)):
    anime_idx, score = sorted_anime[i]
    if score >= similarity_threshold:
      recommended_anime.append(df.iloc[anime_idx]['name'])

  # Limit the number of recommendations to top_n
  return recommended_anime[:top_n] 
```

```python
# Given a target anime, recommend a list of similar anime based on cosine similarity scores
target_anime = "Cross Game" 
recommendations = recommend_anime(target_anime, df, top_n=5)

if recommendations:
  print(f"Anime Recommendations for '{target_anime}':")
  for anime in recommendations:
    print(anime) 
```

```python
# Experiment with different threshold values for similarity scores to adjust the recommendation list size.
target_anime2 = "Kimi no Na wa."

print("Recommendations with similarity threshold 0.6:")
recommendations_06 = recommend_anime(target_anime2, df, top_n=10, similarity_threshold=0.6)
for anime in recommendations_06:
  print(anime)

target_anime3 = "Hajime no Ippo: Rising"

print("\nRecommendations with similarity threshold 0.4:")
recommendations_04 = recommend_anime(target_anime3, df, top_n=10, similarity_threshold=0.4)
for anime in recommendations_04:
  print(anime) 
```

```python
# 4. Evaluation:
# split the data into X and Y variables
X = pd.concat([X_cont_MM, X_cate], axis=1)
print(X.shape)

y = X['type'] 
print(y.shape) 
```

```python
#Impute null values with mode
for col in X.columns:
  X[col].fillna(X[col].mode()[0], inplace=True)
X
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Model fitting
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X, y)

y_pred= model.predict(X) 
y_pred= pd.DataFrame(y_pred)
y_pred.value_counts()
```

```python
# Evaluate using appropriate metrics such as precision, recall, and F1-score.
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y, y_pred, average='micro')
recall = recall_score(y, y_pred, average='micro')
f1 = f1_score(y, y_pred, average='micro')

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1) 
```
