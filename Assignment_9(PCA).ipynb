```python
# 1. Exploratory Data Analysis (EDA):
import pandas as pd
df = pd.read_csv('wine.csv')
df
```

```python
# Basic data exploration
print(df.info())  
```

```python
print(df.describe())
```

```python
# 2. Examine the distribution of features using histograms, box plots, or density plots.
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
df.hist(figsize=(12, 8), bins=15, edgecolor='black')
plt.suptitle('Histograms of Wine Features', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Box plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title('Boxplots of Wine Features', fontsize=16)
plt.xticks(rotation=45)
plt.show()

# Density plots (example for 'Alcohol' feature)
sns.kdeplot(df['Alcohol'])
plt.title('Density Plot of Alcohol Content')
plt.xlabel('Alcohol')
plt.show()
```

```python
# 3. Investigate correlations between features to understand relationships within the data.
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Print the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Wine Features')
plt.show()
```

```python
# 2. Dimensionality Reduction with PCA:
# 1. Standardize the features
from sklearn.preprocessing import StandardScaler

X_cont = df.iloc[:,:]
SS = StandardScaler()
X_cont_SS = SS.fit_transform(X_cont)
X_cont_SS = pd.DataFrame(X_cont_SS)
X_cont_SS.columns = list(X_cont)
X_cont_SS.head()
```

```python
# 2. Implement PCA and determine optimal number of components
from sklearn.decomposition import PCA
import numpy as np

pca = PCA()
pca.fit(X_cont_SS)

# Scree plot
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.title('Scree Plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.show()

# Cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
# 3. Transform the dataset into principal components
pca = PCA(n_components=3)
principal_components = pca.fit_transform(X_cont_SS)

# Create a DataFrame with the principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])
pc_df.head()
```

```python
# 3. Clustering with Original Data:
# 1. Apply K-means clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust n_clusters as needed
kmeans_labels = kmeans.fit_predict(X_cont_SS)

df['KMeans_Cluster'] = kmeans_labels
df['KMeans_Cluster']
```

```python
# 2. Visualize clustering results:
# Using the first two principal components for visualization
plt.scatter(pc_df['PC1'], pc_df['PC2'], c=kmeans_labels, cmap='rainbow')
plt.title('K-Means Clustering Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# 3. Evaluate clustering performance
# Silhouette Score
from sklearn.metrics import silhouette_score
kmeans_silhouette = silhouette_score(X_cont_SS, kmeans_labels)
print("Silhouette Score for K-Means:", kmeans_silhouette)
```

```python
# 4. Clustering with PCA Data:
# 1. Apply K-means clustering to PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca_labels = kmeans_pca.fit_predict(principal_components)

# Add cluster labels to the PCA DataFrame
pc_df['KMeans_PCA_Cluster'] = kmeans_pca_labels
pc_df['KMeans_PCA_Cluster']
```

```python
# 2. Visualize clustering results from PCA-transformed data
plt.scatter(pc_df['PC1'], pc_df['PC2'], c=kmeans_pca_labels, cmap='rainbow')
plt.title('K-Means Clustering Visualization (PCA-Transformed Data)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# 3. Compare clustering results
# Calculate Silhouette Score for K-Means on PCA-transformed data
kmeans_pca_silhouette = silhouette_score(principal_components, kmeans_pca_labels)
print("Silhouette Score for K-Means (PCA-Transformed):", kmeans_pca_silhouette)
```

```python
# Compare with results from the original dataset
print("\nComparison:")
print("Silhouette Score (Original):", kmeans_silhouette)
print("Silhouette Score (PCA-Transformed):", kmeans_pca_silhouette)
```

```python
# 5. Analysis:
# Trade-offs between PCA and Clustering on Original Data:

# Consider the benefits of PCA:
# - Reduced dimensionality can lead to faster computation.
# - May improve clustering performance by removing noise or irrelevant features.

# Consider the potential drawbacks of PCA:
# - Loss of information due to dimensionality reduction.
# - Interpretability of clusters might be affected as principal components are linear combinations of original features.
```
