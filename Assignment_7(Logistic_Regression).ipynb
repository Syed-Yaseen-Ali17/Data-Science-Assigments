```python
# 2. Data Preprocessing:
#a. Handle missing values
import pandas as pd

# Impute missing values
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
# Encode categorical variables:
# Label encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
X_cate = df[['Survived','Name','Sex','Ticket','Cabin','Embarked']]

X_cate['Survived']=LE.fit_transform(X_cate['Survived'])
X_cate['Name']=LE.fit_transform(X_cate['Name'])
X_cate['Sex']=LE.fit_transform(X_cate['Sex'])
X_cate['Ticket']=LE.fit_transform(X_cate['Ticket'])
X_cate['Cabin']=LE.fit_transform(X_cate['Cabin'])
X_cate['Embarked']=LE.fit_transform(X_cate['Embarked'])

X_cate.head() 
```

```python
# Standard scaling
X_cont = titanic_train[['PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]

#Standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_cont_SS = SS.fit_transform(X_cont)
X_cont_SS = pd.DataFrame(X_cont_SS)
X_cont_SS.columns = list(X_cont)
X_cont_SS.head() 
```

```python
# Prepare the features and target variable
X = pd.concat([X_cont_SS, X_cate], axis=1)
print(X.shape)

y = X['Survived']
print(y.shape)
```

```python
from sklearn.impute import SimpleImputer

# Define the imputer
imputer = SimpleImputer(strategy='mean') 
X_imputed = imputer.fit_transform(X) 
X = pd.DataFrame(X_imputed, columns=X.columns) 
X  
```

```python
# 3. Model Building:
# a. Build a logistic regression model using appropriate libraries (e.g., scikit-learn).                                                                                  b. Train the model using the training data.

import pandas as pd
from sklearn.linear_model import LogisticRegression

# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Create a logistic regression model
model = LogisticRegression(multi_class='ovr', max_iter=200)

# b. Train the model using training data
model.fit(X_train, y_train) 
```

```python
# 4. Model Evaluation:
# a. Evaluate the performance of the model on the testing data 
y_pred = model.predict(X_test)
y_pred = pd.DataFrame(y_pred)
print(y_pred.value_counts())

# Get probability estimates for each class
y_proba = model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize

ac= accuracy_score(y_test, y_pred)
print("Accuracy score:", ac.round(2))

rs = recall_score(y_test, y_pred, average='weighted')
print("recall score:", rs.round(2))  

ps = precision_score(y_test, y_pred, average='weighted')
print("precision score:", ps.round(2)) 

f1 = f1_score(y_test, y_pred, average='weighted')
print("f1 score:", f1.round(2)) 

roc_auc = roc_auc_score(y_test, y_proba, average='weighted', multi_class='ovr')
print("roc-auc score:", roc_auc.round(2))  
```

```python
# INTERVIEW QUESTIONS:

# 1. Difference Between Precision and Recall

# Precision:
# - Precision measures the proportion of true positive predictions among all positive predictions made by the model.
# - High precision indicates that when the model predicts a positive class, it is often correct. Precision is crucial when the cost of false positives is high. 

# Recall:
# - Recall measures the proportion of true positive predictions among all actual positive instances in the dataset.
# - High recall means that the model identifies most of the positive instances. Recall is important when the cost of false negatives is high.

# Key Differences:
# - Focus: Precision focuses on the quality of positive predictions (how many of the predicted positives are actually positive), while recall focuses on the completeness of positive predictions (how many of the actual positives are identified).
# - Trade-off: There is often a trade-off between precision and recall. Increasing precision typically reduces recall and vice versa. This is managed using the F1 Score, which is the harmonic mean of precision and recall, to balance the trade-off.
```

```python
# 2. Cross-Validation

# Definition:
# Cross-validation is a technique used to assess the performance and generalizability of a model by dividing the dataset into multiple subsets (folds) and systematically training and validating the model on different subsets.

# Working:
# 1. Partition the Data: The data is divided into k folds. Common values for k are 5 or 10.
# 2. Training and Validation: For each fold, the model is trained on k-1 folds and validated on the remaining fold. This process is repeated k times, each time with a different fold as the validation set.
# 3. Aggregate Results: The performance metrics (such as accuracy, precision, recall, etc.) are averaged over the k iterations to provide a more robust evaluation of the model’s performance.

# Importance in Binary Classification:
# - Performance Estimation: Cross-validation provides a more reliable estimate of a model’s performance compared to a single train-test split. It helps in understanding how well the model generalizes to unseen data.
# - Avoid Overfitting: By training and validating on different subsets of the data, cross-validation helps in detecting overfitting. It ensures that the model’s performance is not overly dependent on a particular subset of the data.
# - Model Selection: It aids in comparing different models or hyperparameters by providing a consistent and unbiased evaluation metric across different folds of the dataset.
# - Reduced Variance: Averaging the performance metrics across multiple folds reduces the variance of the evaluation metric, providing a clearer picture of the model’s effectiveness.
```
