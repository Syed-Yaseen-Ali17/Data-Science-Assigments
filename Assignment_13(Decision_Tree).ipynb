```python
# 3. Feature Engineering:
# Label encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
X_cate = df[['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']]

X_cate['sex']=LE.fit_transform(X_cate['sex'])
X_cate['cp']=LE.fit_transform(X_cate['cp'])
X_cate['fbs']=LE.fit_transform(X_cate['fbs'])
X_cate['restecg']=LE.fit_transform(X_cate['restecg'])
X_cate['exang']=LE.fit_transform(X_cate['exang'])
X_cate['slope']=LE.fit_transform(X_cate['slope'])
X_cate['thal']=LE.fit_transform(X_cate['thal'])

X_cate.head() 
```

```python
# Standard Scaling
X_cont = df[['age', 'trestbps', 'chol', 'thalch', 'oldpeak', 'num']]

from sklearn.preprocessing import StandardScaler 

SS = StandardScaler()
X_cont_SS = SS.fit_transform(X_cont)
X_cont_SS = pd.DataFrame(X_cont_SS)
X_cont_SS.columns = list(X_cont)
X_cont_SS.head()
```

```python
# Check for missing values
df_dropped = df.dropna()

# Impute missing values with the mean (for numerical features)
numerical_cols = ['age', 'trestbps', 'chol', 'thalch', 'oldpeak','num']
for col in numerical_cols:
  df[col] = df[col].fillna(df[col].mean())

# Impute missing values with the mode (for categorical features)
categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal'] 
for col in categorical_cols:
  df[col] = df[col].fillna(df[col].mode()[0])

print(df_dropped.isnull().sum())
print(df.isnull().sum())
```

```python
# Concatenation
X = pd.concat([X_cont_SS, X_cate], axis=1)
print(X.shape)

Y = X['exang']
print(Y.shape)
```

```python
# 4. Decision Tree Classification:
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Implement a Decision Tree Classification model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

# Train the model on the training set
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)
y_pred = pd.DataFrame(y_pred)
y_pred.value_counts() 
```

```python
# Evaluate the model's performance
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("ROC-AUC:", roc_auc)
```

```python
# 5. Hyperparameter Tuning:
# Perform hyperparameter tuning to optimize the Decision Tree model. Experiment with different hyperparameters such as maximum depth, minimum samples split, and criterion.

from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}

# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=param_grid, cv=5, scoring='accuracy')

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

best_model = grid_search.best_estimator_

y_pred_best = best_model.predict(X_test)

# Evaluate the best model's performance
accuracy_best = accuracy_score(y_test, y_pred_best)
precision_best = precision_score(y_test, y_pred_best)
recall_best = recall_score(y_test, y_pred_best)
f1_best = f1_score(y_test, y_pred_best)
roc_auc_best = roc_auc_score(y_test, y_pred_best)

print("Best Model Performance:")
print("Accuracy:", accuracy_best)
print("Precision:", precision_best)
print("Recall:", recall_best)
print("F1-score:", f1_best)
print("ROC-AUC:", roc_auc_best)
```

```python
# INTERVIEW QUESTIONS:

# 1. Decision tree models have several hyperparameters that significantly impact their performance.
# Here are some of common hyperparameters and how they affect a decision treeâ€™s behavior:

# 1. Max Depth
 # - The maximum number of levels or layers in the tree. It limits how deep the tree can grow.
 # - Controlling the depth helps prevent overfitting. A very deep tree can model the training data too closely, capturing noise and leading to overfitting.

# 2. Min Samples Split
# - The minimum number of samples required to split an internal node. If a node has fewer samples than this value, it wonâ€™t be split.
# - Increasing this value prevents the model from learning overly specific rules that apply to very few samples, which can help prevent overfitting.

# 3. Min Samples Leaf
 # - The minimum number of samples required to be at a leaf node. This parameter sets a threshold to ensure that leaf nodes contain a minimum number of samples.
 # - Setting this value helps smooth the model by ensuring that leaf nodes represent a sufficient number of samples.

# 4. Max Features
 # - The number of features to consider when looking for the best split. It can be an integer, a float (representing a fraction of total features), or a string ('auto', 'sqrt', 'log2').
 # - Restricting the number of features can help in reducing overfitting by introducing randomness into the model, which can lead to better generalization.

# 5. Max Leaf Nodes
 # - The maximum number of leaf nodes in the tree.
 # - Limiting the number of leaf nodes can help prevent overfitting by restricting the complexity of the model.

# 6. Criterion
 # - The function to measure the quality of a split. Common criteria are "gini" for Gini impurity and "entropy" for information gain.
 # - The choice of criterion affects how splits are determined. "Gini" tends to be faster and less computationally intensive, while "entropy" can provide slightly better performance in some cases but at the cost of additional computation.

# Summary of Impacts:
 # - Overfitting vs. Underfitting: Hyperparameters like max depth, min samples split, and min samples leaf are crucial in balancing between overfitting and underfitting.
 # - Complexity and Generalization: Parameters that limit tree complexity (e.g., max features, max leaf nodes) help in building models that generalize better to unseen data.
 # - Training Time: Parameters like max depth and max features also influence the training time. Deeper trees or trees with more features can be slower to train.

# Tuning these hyperparameters involves finding the right balance between model complexity and performance,
# often using techniques like cross-validation to evaluate how different settings affect the modelâ€™s ability to generalize to new data.
```

```python
# 2. Label encoding and one-hot encoding

# Label encoding:
# Definition: Label encoding involves converting each category in a categorical feature into a unique integer. Each category is assigned a specific integer value.

# Working:
# 1. Assign Integer Labels: Each unique category in the feature is assigned an integer value.
# 2. Replace Categories: The categorical feature values are replaced with their corresponding integer labels.

# Example:
# - Original Categories: ["Red", "Green", "Blue"]
# - Encoded Values: [0, 1, 2]

# One-Hot Encoding:
# Definition: One-hot encoding converts each category into a binary vector.
# Each category is represented by a vector with a length equal to the number of unique categories, where only one element is set to 1 and all others are set to 0.

# Working:
# 1. Create Binary Columns: For each unique category, create a new binary feature (column).
# 2. Set Values: In the new binary columns, set the value to 1 for the column corresponding to the category and 0 for all other columns.

# Example:
# - Original Categories: ["Red", "Green", "Blue"]
# - One-Hot Encoding:
  # - "Red" â†’ [1, 0, 0]
  # - "Green" â†’ [0, 1, 0]
  # - "Blue" â†’ [0, 0, 1]

# Choosing between these methods depends on the nature of the categorical feature and the requirements of the machine learning algorithm.
```
